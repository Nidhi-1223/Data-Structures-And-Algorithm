## Traversal in a Tree

1. Depth First Search 
    - PreOrder Traversal
    - PostOrder Traversal
    - InOrder Traveral

2. Breadth First Search
    - Level Order Traversal

### Pre-Order Traversal

Time Complexity = O(n)

Space Complexity = O(n)

Whenever a function is called recursively it inserts elements in a stack to remember it's progress. ( **We're using stack because this is a DFS algorithm** )

### In-Order Traversal

Time Complexity = O(n)

Space Complexity = O(n)

### Post-Order Traversal

Time Complexity = O(n)

Space Complexity = O(n)

### Level-Order Traversal
We'll be using a queue to store elements

```python
# use the leftchild, rightchild method for traversal
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.leftChild = None
        self.rightChild = None

def preOrderTraversal(rootNode):
    # root -> left SubTree -> right SubTree
    if rootNode:
        print(rootNode.data)
        preOrderTraversal(rootNode.leftChild)
        preOrderTraversal(rootNode.rightChild)

def inOrderTraversal(rootNode):
    # left SubTree -> root -> right SubTree
    if rootNode:
        inOrderTraversal(rootNode.leftChild)
        print(rootNode.data)
        inOrderTraversal(rootNode.rightChild)

def postOrderTraversal(rootNode):
    # left SubTree -> right SubTree -> root
    if rootNode:
        postOrderTraversal(rootNode.leftChild)
        postOrderTraversal(rootNode.rightChild)
        print(rootNode.data)

newBT = TreeNode('Drinks')

lev01_leftchild = TreeNode('Hot')
lev01_rightchild = TreeNode('Cold')

lev12_leftChild = TreeNode('Tea')
lev12_rightChild = TreeNode('Coffee')

lev22_leftChild = TreeNode('Alcoholic')
lev22_rightChild = TreeNode('Non-Alcoholic')

newBT.leftChild = lev01_leftchild
newBT.rightChild = lev01_rightchild

lev01_leftchild.leftChild = lev12_leftChild
lev01_leftchild.rightChild = lev12_rightChild

lev01_rightchild.leftChild = lev22_leftChild
lev01_rightchild.rightChild = lev22_rightChild

print('-------Pre-order Traversal--------')
preOrderTraversal(newBT)

print('-------In-order Traversal--------')
inOrderTraversal(newBT)

print('-------Post-order Traversal--------')
postOrderTraversal(newBT)
```

## Inserting a Node in a Tree

For inserting a node in a tree we perform level order traversal and insert it in the first vacant position found.

```python
def inserting(rootNode, newNode):
    if not rootNode:
        print('The Binary tree doesnt exist')
    else:
        custQueue = DynamicQueue()
        custQueue.enQueue(rootNode)
        while not(custQueue.isEmpty()):
            dequeued_element = custQueue.deQueue()
            if (dequeued_element.value.leftChild is not None):
                custQueue.enQueue(dequeued_element.value.leftChild)
            else:
                dequeued_element.value.leftChild = newNode
                return 'Element inserted'
            
            if (dequeued_element.value.rightChild is not None):
                custQueue.enQueue(dequeued_element.value.rightChild)
            else:
                dequeued_element.value.rightChild = newNode
                return 'Element inserted'


newNode = TreeNode('new node')
inserting(newBT, newNode)
levelOrderTraversal2(newBT)
```

## Deleting a Node from Tree
To delete a node, we replace its value with the deepest element and then delete the deepest element

```python
#helper function #1
def getDeepestNode(rootNode):
    if not rootNode:
        return 
    else:
        custQueue = DynamicQueue()
        custQueue.enQueue(rootNode)
        while not(custQueue.isEmpty()):
            dequeued_element = custQueue.deQueue()
            if (dequeued_element.value.leftChild is not None):
                custQueue.enQueue(dequeued_element.value.leftChild)
            if (dequeued_element.value.rightChild is not None):
                custQueue.enQueue(dequeued_element.value.rightChild)
        deepestNode = dequeued_element.value
        return deepestNode

# helper function #2
def deleteDeepestNode(rootNode,deepestNode):
    if not rootNode:
        return 
    else:
        custQueue = DynamicQueue()
        custQueue.enQueue(rootNode)
        while not(custQueue.isEmpty()):
            dequeued_element = custQueue.deQueue()
            if dequeued_element.value is deepestNode:
                print(dequeued_element.value.data)
                dequeued_element.value = None
                return
            if dequeued_element.value.leftChild:
                if dequeued_element.value.leftChild is deepestNode:
                    dequeued_element.value.leftChild = None
                    return 
                else:
                    custQueue.enQueue(dequeued_element.value.leftChild)

            if dequeued_element.value.rightChild:
                if dequeued_element.value.rightChild is deepestNode:
                    dequeued_element.value.rightChild = None
                    return 
                else:
                    custQueue.enQueue(dequeued_element.value.rightChild)

def deleteNode(rootNode, to_delete):
    if not rootNode:
        return 
    else:    
        custQueue = DynamicQueue()
        custQueue.enQueue(rootNode)

        while not(custQueue.isEmpty()):
            dequeued_element = custQueue.deQueue()

            if dequeued_element.value.data == to_delete:
                dNode = getDeepestNode(rootNode)
                dequeued_element.value.data = dNode.data

                deleteDeepestNode(rootNode, dNode)
                return 'The node has been successfully deleted'
            
            if (dequeued_element.value.leftChild is not None):
                custQueue.enQueue(dequeued_element.value.leftChild)

            if (dequeued_element.value.rightChild is not None):
                custQueue.enQueue(dequeued_element.value.rightChild)
                
        return 'failed to delete'  


# dNode = getDeepestNode(newBT)
# deleteDeepestNode(newBT, dNode)
# deleteNode(newBT, 'Coffee')
levelOrderTraversal2(newBT)
```

# Tree implementation using Python Lists

```python
class BinaryTree:
    def __init__(self, size):
        self.customList = size * [None]
        self.lastUsedIndex = 0
        self.maxSize = size
    def insertNode(self, value):
        if self.lastUsedIndex+1 == self.maxSize:
            return 'The Binary Tree is full'
        else:
            self.customList[self.lastUsedIndex + 1] = value
            self.lastUsedIndex += 1
            return 'Node added successfully!'
    
    def searchNode(self, value):
        for i in range(len(self.customList)):
            if self.customList[i] == value:
                return 'Found it!'
        return 'Not Found'
    
    def preOrderTraversal(self,index):
        if index > self.lastUsedIndex: # because everytime an element is inserted we increment the lastUsedIndex by 1
            return 
        print(self.customList[index])
        self.preOrderTraversal(index*2)
        self.preOrderTraversal(index*2 + 1)

    def inOrderTraversal(self,index):
        if index > self.lastUsedIndex: # because everytime an element is inserted we increment the lastUsedIndex by 1
            return 
        self.inOrderTraversal(index*2)
        print(self.customList[index])
        self.inOrderTraversal(index*2 + 1)
    
    def postOrderTraversal(self,index):
        if index > self.lastUsedIndex: # because everytime an element is inserted we increment the lastUsedIndex by 1
            return 
        self.postOrderTraversal(index*2)
        self.postOrderTraversal(index*2 + 1)
        print(self.customList[index])
    
    def levelOrderTraversal(self,index):
        for i in range(index, self.lastUsedIndex+1):
            print(self.customList[i])

    def deleteNode(self, value):
        if self.lastUsedIndex == 0:
            return 'There is no node to delete bitch'
        self.deepestNode = self.customList[self.lastUsedIndex]
        for i in range(1, self.lastUsedIndex+1):
            if self.customList[i] == value:
                self.customList[i] = self.deepestNode
                self.customList.pop(self.lastUsedIndex)
                # self.customList[self.lastUsedIndex] = None       # alternate way to pop the element
                self.lastUsedIndex -= 1
                return 'The node has been deleted successfully'

    def deleteBinaryTree(self):
        self.customList = None
        return 'Binary tree deleted'



listBT = BinaryTree(8)
listBT.insertNode('Drinks')
listBT.insertNode('Hot')
listBT.insertNode('Cold')
listBT.insertNode('Tea')
listBT.insertNode('Coffee')
listBT.insertNode('Non-Alcoholic')
listBT.insertNode('Alcoholic')

# listBT.searchNode('Cold')

# print('Preorder traversal -------------------')
# listBT.preOrderTraversal(1)
# print('Inorder traversal -------------------')
# listBT.inOrderTraversal(1)
# print('Postorder traversal -------------------')
# listBT.postOrderTraversal(1)
# print('Levelorder traversal -------------------')
# listBT.levelOrderTraversal(1)

listBT.deleteNode('Tea')
listBT.levelOrderTraversal(1)
```
